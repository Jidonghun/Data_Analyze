## 내 생일의 기온 변화를 그래프로 그리기

````python
import csv

f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)

max_temp_list =[]

for row in data:
  if row[-1] != '':                         # 누락된 데이터가 아닐 경우(데이터가 있을 경우)
    max_temp_list.append(float(row[-1]))    # row[-1]은 문자열 데이터임으로 float() 함수를 사용하여 숫자형 데이터로 변환 후 저장

f.close()

print(max_temp_list)
print( len(max_temp_list) )
````

````python
import matplotlib.pyplot as plt

plt.plot(max_temp_list, 'blue')
plt.show()
````

````python
plt.figure(figsize=(16,4))
plt.plot(max_temp_list, 'blue')
plt.show()
````

### Quiz
매년 생일 날의 최고 기온과 최저 기온을 하나의 그림(figure)에 나타내는 코드를 작성하세요.

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)
max_temp_list =[]
min_temp_list = []

for row in data:
  if row[-1] != '' and row[-2] != '':                                                             # 누락된 데이터가 아닐 경우(데이터가 있을 경우)
    if row[0].split('-')[1] == '08' and row[0].split('-')[2] =='27':
      max_temp_list.append(float(row[-1]))                                      # row[-1]은 문자열 데이터임으로 float() 함수를 사용하여 숫자형 데이터로 변환 후 저장
      min_temp_list.append(float(row[-2]))

f.close()

plt.figure(figsize=(16, 4))
plt.plot(max_temp_list, 'r', label='max. temp')
plt.plot(min_temp_list, 'b', label='min. temp')
plt.title('max, min temp of every year Augest 27')
plt.ylabel('temp')
plt.legend()
plt.show()
````

## 기온 데이터를 히스토그램으로 표현하기

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)
max_temp_list =[]

for row in data:
  if row[-1] != '':
    max_temp_list.append(float(row[-1]))

f.close()

plt.hist(max_temp_list, bins=100, color='r')
plt.show()
````

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)
max_temp_list =[]

for row in data:
  if row[-1] != '':
    ymd = row[0].split('-')
    if ymd[1] == '08':
      max_temp_list.append(float(row[-1]))

f.close()

plt.hist(max_temp_list, bins=100, color='r')
plt.show()
````

````python
import csv
f = open('서울.csv', encoding='cp949')
data = csv.reader(f)
next(data)
max_temp_Alist =[]
max_temp_Blist =[]

for row in data:
  if row[-1] != '':
    ymd = row[0].split('-')
    if ymd[1] == '08':
      max_temp_Alist.append(float(row[-1]))
    if ymd[1] == '01':
      max_temp_Blist.append(float(row[-1]))

f.close()

plt.hist(max_temp_Alist, bins=100, color='r')
plt.hist(max_temp_Blist, bins=100, color='b')
plt.title('1월, 8월 최고 기온 분포')
plt.show()
````
