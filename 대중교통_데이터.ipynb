```python
# Myself
# 수정필요

import csv
import matplotlib.pyplot as plt
import koreanize_matplotlib

f = open('subway.csv')
data = csv.reader(f)
next(data)

name = []
pay = []
non = []

for row in data:
  for idx in range(4,8):
    row[idx] = int(row[idx])
  # print(row)
  pay.append(row[4])
  non.append(row[6])
  name.append(row[3])

# print(pay)
# print(non)
f.close()

payrates = []

for i in range(len(pay)):
  try:
    payrate = round(pay[i]/non[i])
  except:
    print('0')
  else:
    payrates.append(payrate)

# print(name)
print("최대값:", max(payrates), payrates.index(55), "최소값:", min(payrates), payrates.index(0))
print("유임 승차율이 최대인 역은", name[72], "유임승차율이 최소인 역은", name[325])
print(payrates)
plt.figure(figsize=(16, 4))
# plt.plot(payrates, 'r')
plt.bar(name[:307], payrates[:307])
plt.title('유임승차율')
plt.ylabel('payrates')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import koreanize_matplotlib

f = open('subway.csv')
data = csv.reader(f)
next(data)

name = []
rate = []

for row in data:
  for idx in range(4,8):
    row[idx] = int(row[idx])
  if row[6] != 0:
    rate.append(row[4]/row[6])
    name.append(row[3])

f.close()

for i in range(len(name)):
  # print('{} : {:.2f}%'.format(name[i], rate[i]*100))
  if rate[i] == max(rate):
    print(name[i], ':', rate[i]*100, '%')
```
